```python
sum_30to50 = pickle.load(open("summary_1930to1950.pkl", "rb"))
sum_51to60 = pickle.load(open("summary_1951to1960.pkl", "rb"))
sum_61to70 = pickle.load(open("summary_1961to1970.pkl", "rb"))
summaries = sum_30to50 + sum_51to60 + sum_61to70
print(summaries[3])
```

```python
years = [year[0] for year in summaries]
mean_tmp = [year[1] for year in summaries]
min_tmp = [year[2] for year in summaries]
max_tmp = [year[3] for year in summaries]

plt.plot(years, max_tmp, 'r', label="Maximum")
plt.plot(years, mean_tmp, 'g', label="Average")
plt.plot(years, min_tmp, 'b', label="Minimum")

plt.title("Global Temperatures from GSOD")
plt.xlabel("Year")
plt.ylabel("Temperature in Fahrenheit")
plt.legend()

plt.show()
```
